```python
import matplotlib.pyplot as plt
```

Usaremos o módulo matplotlib.pyplot, o pyplot nos permite construir gráficos passo a passo


```python
# dados

x = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
y = [300, 543, 1075, 2862, 5979, 10287, 14956] 

# criando um gráfico de linhas 
plt.plot(x,y, color='green', marker='o', linestyle='solid')

# adicionando um título
plt.title('Título')

# nomeando eixos 
plt.ylabel('ordenada')
plt.xlabel('abcissa')

# exibindo
plt.show()
```

Gráfico de barras


```python
# gráfico de barras são muito úteis para visualizar como alguma quantidade varia em
# um número discreto de itens

Filmes = ['Annie Hall', 'Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscar = [5, 11, 3, 8, 10]

# criando um gráfico de barras
plt.bar(Filmes, num_oscar)

plt.title('Filmes')
plt.ylabel('Número de oscars')

plt.show()
```

```python
# Podemos também criar um histograma que fornece uma visualização da distribuição
# dos valores


Notas = [83, 95, 91, 87, 60, 75, 35, 82, 55, 70, 66, 61, 40, 90, 65, 71, 55, 84, 60, 59, 31, 92]

# criando um histograma
plt.hist(Notas, bins=15)
plt.show()
```

Gráfico de linhas

```python
# Gráficos de linhas são bons para mostrar tendências

A = [1, 2, 4, 8, 16, 32, 64, 128, 256]
B = [256, 128, 64, 32, 16, 8, 4, 2, 1]
erro = [x+y for x, y in zip(A, B)]
xs = [i for i, _ in enumerate(A)]

# Podemos fazer várias chamadas para plt.plot

plt.plot(xs, A, 'g-', label='A') # g- para linha sólida verda
plt.plot(xs, B, 'r-', label='B') # r- para linha sólida vermelha
plt.plot(xs, erro, 'b:', label='erro') # b: para linha tracejada azul

# como cada série já tem um rótulo atribuido, precisamos criar uma legenda para
# mostrar estes rótulos

plt.legend(loc=9) # loc=9 significa "centralizado no topo"
plt.show()  
```

Gráficos de dispersão

```python
# É a opção correta para demonstrar as relações entre pares de conjuntos de dados
# exemplo:
# relação entre número de amigos adicionados e minutos passados em uma rede social

amigos = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutos = [175, 170, 205, 120, 220, 130, 105, 145, 190]
label = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(amigos, minutos)

# rotulando cada ponto 

for label, amigos_count, minutos_count in zip(label, amigos, minutos):
    plt.annotate(label, 
        xy=(amigos_count, minutos_count), # colocando rótulo no ponto xy
        xytext=(5,-5),                    # levemente deslocado
        textcoords='offset points')

plt.xlabel('Amgos')
plt.ylabel('Minutos')
plt.show()
```

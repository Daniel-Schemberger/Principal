```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def map_1(n,x0):
    ne = [i for i in range(n+1)]
    x = [0]*(n+1)
    x[0] = x0
#x_n+1 = 2x_n mod 1
    for i in range(n):
        x[i+1] = 2*x[i]%1

    return ne, x

def map_2(n,x0):
    ne = [i for i in range(n+1)]

    x = [0]*(n+1)
    x[0] = x0
#x_n+1 = 4x_n(1-x_n)
    for i in range(n):
        x[i+1] = 4*x[i]*(1-x[i])

    return ne, x

def map_3(n,x0):
    ne = [i for i in range(n+1)]
    x = [0]*(n+1)
    x[0] = x0
#x_n+1 = 3.1x_n mod 1
    for i in range(n):
        x[i+1] = (3.1*x[i])%1

    return ne, x
```

```python
def plot():
    n=1000

#map_1
    fig, ax = plt.subplots() 
    res1 = map_1(n,0.54321)
    res2 = map_1(n,0.54322)

    ax.plot(res1[0], res1[1], '.')
    ax.plot(res2[0], res2[1], '.')
    ax.legend(('x_0 = 0.54321','x_0 = 0.54322'))
    ax.set_title(" x_{n+1} = 2x_n  mod 1 ")
    ax.set_xlabel('n')
    ax.set_ylabel('x_n')

#map_3
    fig, ax = plt.subplots()
    res3 = map_3(n,0.54321)
    res4 = map_3(n,0.54322)

    ax.plot(res3[0],res3[1],'.')
    ax.plot(res4[0],res4[1],'.')
    ax.legend(('x_0 = 0.54321','x_0 = 0.54322'))
    ax.set_title(" x_{n+1} = 3.1x_n  mod 1 ")
    ax.set_xlabel('n')
    ax.set_ylabel('x_n')

#map_2
    fig, ax = plt.subplots()
    res5 = map_2(n,0.54321)
    res6 = map_2(n,0.54322)
    plt.figure(figsize=(10,10))

    ax.plot(res5[0],res5[1],'.')
    ax.plot(res6[0],res6[1],'.')
    ax.legend(('x_0 = 0.54321','x_0 = 0.54322'))
    ax.set_title("x_{n+1} = 4x_n(1-x_n)")
    ax.set_xlabel('n')
    ax.set_ylabel('x_n')
    
    plt.show()
```

```python
plot()
```

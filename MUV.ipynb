Bibliotecas:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sympy import Symbol
from sympy import plot
from math import sqrt
```

Imput das constantes:

```python
#Em S.I

x0 = float(input('Posição inicial: '))
v0 = float(input('Velocidade inicial: '))
a = float(input('Aceleração: '))
```

Definições e gráfico da função horária da posição:

```python
inter_tempo =10
dt = 0.1
n = int(round(inter_tempo/dt))
t = np.linspace(0,5,n) 

x = x0+v0*t+0.5*a*t**2

plt.plot(t, x, label='Posição em metros')
plt.title('Posição em função do tempo')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.show()
```

Variável simbólica e funções da velocidade e aceleração:

```python
t = Symbol('t')
x = x0+v0*t+0.5*a*t**2
v=x.diff(t)   # dx/dt
a = v.diff(t) # dv/dt
print(f"Posição: {x}")
print(f"Velocidade: {v}")
print(f"Aceleração: {a}")
```

Gráfico da velocidade em função do tempo:

```python
vp = plot(v, (t, 0, 5), 
        title= 'Velocidade em função do tempo',
        xlabel= 't', ylabel= 'V(t)',
        axis_center= (0,0),  ylim= (-10,10))
```

Gráfico da aceleração em função do tempo:

```python
ap = plot(a, (t, 0, 5), 
        title= 'Aceleração em função do tempo',
        xlabel= 't', ylabel= 'a(t)',
        axis_center= (0,0), ylim= (-10,10))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bif(n):

    f = lambda r,x : r*x*(1-x)
    x0, R, X = 0.1, [], []

    for r in np.linspace(0, 4, n):
        x=x0
        for j in range(n):
            x = f(r,x)
        for k in range(n):
            x = f(r,x)
            R.append(r)
            X.append(x)

    plt.figure(figsize=(15,15))
    plt.xlabel('r')
    plt.ylabel('x')
    plt.xlim(1,4)
    plt.ylim(0,1)
    plt.grid(True)
    plt.plot (R,X, '.',markersize=0.5)
    plt.title('Bifurcao do mapa f(x) = rx(x-1)')
    plt.show()


bif(1000)
```

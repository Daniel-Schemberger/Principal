```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
import random
```

```python
p = [0.05, 0.95]
x0 = 2
s = [x0]

rn = np.random.random(1000)
down = rn < p[0]
up = rn > p[1]

for i, j in zip(down, up):
    d = i and p[-1] > 0
    u = j and p[-1] < 5
    p.append(p[-1] - d + u)
plt.plot(p)
plt.show()
```

```python
n = 1000
x = np.zeros(n)
y = np.zeros(n)

for i in range(1, n):
    val = random.randint(1,4)
    if val == 1:
        x[i] = x[i-1] + 1
        y[i] = y[i-1]
    elif val == 2:
        x[i] = x[i-1]
        y[i] = y[i-1] + 1
    elif val == 3:
        x[i] = x[i-1] - 1
        y[i] = y[i-1] 
    else:
        x[i] = x[i-1] 
        y[i] = y[i-1] -1

pylab.title('Caminhada aleat√≥ria 2d')
pylab.plot(x,y)
pylab.show()
```

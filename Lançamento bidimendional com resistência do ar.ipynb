Bibliotecas:

```python
import numpy as np
from scipy.integrate import odeint
from matplotlib import pyplot as plt
import math
```

Imputs das condições iniciais:

```python
x0 = float(input('Posição inicial (x): '))
y0 = float(input('Posição inicial (y): '))
vx0 = float(input('velocidade inicial inicial (x): '))
vy0 = float(input('velocidade inicial inicial (y): '))
```

Parâmetros:

```python

c = 0.5 # parâmetro de arraste
p = 1   # densidade do meio

r = 2  # raio do corpo
m = 5   # massa do corpo
A = math.pi*r**2 # area de sessão transversal

g = 9.8 #aceleração da gravidade
k = round(0.5*(c*p*A/m),5) #constante de arrasto
```

Equações do movimento com arraste:

```python
#Movimento em y
def edoy(y,t):
    return (y[1], -g - k*y[1])

#Movimento em x
def edox(x, t):
    return (x[1], -k*x[1])
```

Armazenando condições iniciais e intervalo de tempo:

```python
yci = [y0, vy0]
xci = [x0, vx0]
ts = np.linspace(1, 3.5, 25)
```

Solução das EDO's:

```python
us = odeint(edoy, yci, ts)
us2 = odeint(edox, xci, ts)
ys = us[:, 0]
xs = us2[:, 0]
```

Gráfico da posição em x pelo tempo:

```python
plt.plot(ts, ys, '-', label='y(t)')
plt.xlabel('tempo')
plt.ylabel('posição')
plt.title('Posição (y) em função do tempo')
plt.ylim(0, 6)
plt.legend()
plt.grid(True)

plt.show()
```

Gráfico da posição em y pelo tempo:

```python
plt.plot(ts, xs, '-', color='#17a589', label='x(t)')
plt.xlabel('tempo')
plt.ylabel('posição')
plt.title('Posição (x) em função do tempo')
plt.legend()
plt.grid(True)

plt.show()
```

Gráfico da trajetória:

```python
plt.plot(xs, ys, '-')
plt.xlabel('posição em x')
plt.ylabel('posição em y')
plt.title('Trajetória')
plt.ylim(0,6)
plt.grid(True)

plt.show()
```
